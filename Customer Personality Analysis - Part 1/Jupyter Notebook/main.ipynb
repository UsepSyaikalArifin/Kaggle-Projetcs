```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = new_dataset.columns[new_dataset.dtypes != "object"].to_list()
numerical_columns.remove("jumlah anak")
numerical_columns.remove("senioritas")
numerical_columns_outlier = numerical_columns
fig, axs = plt.subplots(len(numerical_columns_outlier), 1, figsize=(12, 6))
for i,column in enumerate(numerical_columns_outlier):
    color = sns.color_palette("Set{}".format(i+1))
    sns.boxplot(data=new_dataset, x=column, palette=color, ax=axs[i])
    axs[i].set_title("Boxplot {}".format(column), fontsize=20, fontweight="bold")
    axs[i].set_xlabel(column, fontsize=15)
plt.tight_layout()
plt.show()
```

```python
Q1 = new_dataset[numerical_columns_outlier].quantile(0.25)
Q3 = new_dataset[numerical_columns_outlier].quantile(0.75)
IQR = Q3 - Q1

batas_minimal = (Q1 - 1.5 * IQR)
batas_maksimal = (Q3 + 1.5 * IQR)

no_outlier_new_dataset = new_dataset[~((new_dataset[numerical_columns_outlier] < batas_minimal) | (new_dataset[numerical_columns_outlier] > batas_maksimal)).any(axis=1)]

display(no_outlier_new_dataset.head(10))
```

## Mengubah Data Kolom tipe object -> Numerik(int64)

```python
from sklearn.preprocessing import LabelEncoder

LE_kolom = ["pendidikan", "status", "anak"]
df_encoder = fix_dataframe[LE_kolom].copy()
for column in LE_kolom:
    LE = LabelEncoder()
    df_encoder[column] = LE.fit_transform(df_encoder[column])
    print(LE.classes_)
    print(sorted(df_encoder[column].unique()))

display(df_encoder.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_kolom = ["usia", "senioritas", "total konsumsi"]
df_minmax = fix_dataframe[minmax_kolom].copy()
minmax_scaller = MinMaxScaler()
df_minmax[minmax_kolom] = minmax_scaller.fit_transform(df_minmax[minmax_kolom])

display(df_minmax.head())
```

## Membuat Dataset Modelling

```python
dataset_modelling = pd.DataFrame({
                                    "pendidikan" : df_encoder["pendidikan"],
                                    "status" : df_encoder["status"],
                                    "anak" : df_encoder["anak"],
                                    "jumlah anak" : fix_dataframe["jumlah anak"],
                                    "usia" : df_minmax["usia"],
                                    "senioritas" : df_minmax["senioritas"],
                                    "total konsumsi" : df_minmax["total konsumsi"]
                                })
display(dataset_modelling.head())
```

```python
from kmodes.kmodes import KModes
from kmodes.kprototypes import KPrototypes

# cost = {}
# for k in range(2, 20):
#     kproto = KPrototypes(n_clusters=k, random_state=0)
#     kproto.fit_predict(dataset_modelling, categorical=[0, 1, 2, 3])
#     cost[k] = kproto.cost_

kproto = KPrototypes(n_clusters=3, random_state=0)
kproto = kproto.fit(dataset_modelling, categorical=[0, 1, 2, 3])
clusters = kproto.predict(dataset_modelling, categorical=[0, 1, 2, 3])

display(clusters)
```

```python
fix_dataframe["Kluster"] = clusters

display(fix_dataframe.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

countplot_kolom = ["pendidikan","status", "anak", "jumlah anak"]
fig, axs = plt.subplots(len(countplot_kolom), 1, figsize=(15, 15))
for i, column in enumerate(countplot_kolom):
    sns.countplot(data=fix_dataframe, x="Kluster", hue=column, ax=axs[i])
    axs[i].set_title("Distribusi Data {}".format(column), fontsize=20, fontweight="bold")
    axs[i].set_xlabel(column, fontsize=15)
    for p in axs[i].patches:
        axs[i].annotate(
                        "%0.2f"%(p.get_height()),
                        xy=(p.get_x() + p.get_width()/2, p.get_height()),
                        ha="center",
                        va="center",
                        xytext=(0, 10),
                        textcoords="offset points"
                    )
    sns.despine(left=True, top=True, right=True)
    axs[i].axes.yaxis.set_visible(False)
plt.tight_layout()
plt.show()
```
