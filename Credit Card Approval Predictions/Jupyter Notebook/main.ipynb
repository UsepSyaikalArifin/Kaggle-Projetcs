```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

from IPython.display import display
```

```python
applications = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\Kaggle\Credit Card Approval Predictions\Dataset\application_record.csv")
print("Shape of dataset = %d Baris & %d Kolom"%applications.shape)
print("Informasi applications : ")
display(applications.info())
print("Statistik Dasar : ")
display(applications.describe())
display(applications.head())
```

```python
credits = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\Kaggle\Credit Card Approval Predictions\Dataset\credit_record.csv")
print("Shape of dataset = %d Baris & %d Kolom"%credits.shape)
print("Informasi credits : ")
display(credits.info())
print("Statistik Dasar : ")
display(credits.describe())
display(credits.head())
```

```python
dataset = pd.merge(applications, credits, left_on="ID", right_on="ID")
dataset = dataset.sort_values(by="ID")
display(dataset.head())
```

```python
rename_columns = {"ID" : "id",
                 "CODE_GENDER" : "jenis kelamin",
                 "FLAG_OWN_CAR" : "punya mobil ?",
                 "FLAG_OWN_REALTY" : "punya properti ?",
                 "CNT_CHILDREN" : "jumlah anak",
                 "AMT_INCOME_TOTAL" : "total pemasukan",
                 "NAME_INCOME_TYPE" : "sumber pemasukan",
                 "NAME_EDUCATION_TYPE" : "status pendidikan",
                 "NAME_FAMILY_STATUS" : "status pernikahan",
                 "NAME_HOUSING_TYPE" : "status tempat tinggal",
                 "DAYS_BIRTH" : "waktu lahir",
                 "DAYS_EMPLOYED" : "waktu awal credits",
                 "FLAG_MOBIL" : "punya mobile phone ?",
                 "FLAG_WORK_PHONE" : "punya Telepon Rumah",
                 "FLAG_PHONE" : "punya HP ?",
                 "FLAG_EMAIL" : "punya email ?",
                 "OCCUPATION_TYPE" : "jenis pekerjaan",
                 "CNT_FAM_MEMBERS" : "jumlah keluarga",
                 "MONTHS_BALANCE" : "cek bulan",
                 "STATUS" : "status pelanggan"}

dataset = dataset.rename(rename_columns, axis=1)
display(dataset.head(5))
```

```python
idn_dataset = dataset.copy()
```

```python
print("Shape of dataset = %d Baris & %d Kolom"%idn_dataset.shape)
print("Informasi dataset : ")
display(idn_dataset.info())
print("Mengecek Missing Value : ")
display(dataset.isnull().sum())
```

## Membuang COlumns "jenis pekerjaa"

```python
idn_dataset = idn_dataset.drop(["jenis pekerjaan"], axis=1)
```

```python
print("Shape of dataset = %d Baris & %d Kolom"%idn_dataset.shape)
print("Informasi dataset : ")
display(idn_dataset.info())
print("Mengecek Missing Value : ")
display(idn_dataset.isnull().sum())
```

## Menghitung Komposisi Data Kategorikal Tipe Object pada masing-masing Kolom Menggunkan Countplot

```python
columns_categorical_object = idn_dataset.columns[idn_dataset.dtypes == "object"].to_list()

fig, axs = plt.subplots(len(columns_categorical_object), 1, figsize=(12, 40))
color = sns.color_palette("bright")

for i, column in enumerate(columns_categorical_object):
    sns.countplot(data=idn_dataset, x=column, ax=axs[i], palette=color)
    axs[i].set_title("Komposisi Data {}".format(column), fontsize=20, fontweight="bold")
    axs[i].set_xlabel("{}".format(column), fontsize=15)
    axs[i].set_ylabel("{}".format(column), fontsize=15)
    for p in axs[i].patches:

        height=p.get_height()
        width=p.get_width()
        x=p.get_x()
        y=p.get_y()

        axs[i].annotate(
                        height,
                        xy=(x + width/2, height),
                        ha="center",
                        va="center",
                        xytext=(0, 10),
                        textcoords="offset points"
                       )
sns.despine(left=True, top=True, right=True)
plt.tight_layout()
plt.show()
```
