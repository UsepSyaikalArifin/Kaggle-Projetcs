```python
import pandas as pd
from IPython.display import display

dataset = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv")

print("Shape of dataset : %d Baris & %d Kolom"%dataset.shape)
print("Informasi dataset : ")
display(dataset.info())
print("Mengecek Missing Value : ")
display(dataset.isnull().sum())
print("Statistik Dasar : ")
display(dataset.describe().round(3))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(1, 1, figsize=(12, 7))
sns.barplot(data=dataset, x="Department /Ministry", y="Fund allotted(in ₹crores)", ax=axs)
axs.set_xticklabels(axs.get_xticks(), rotation=75)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

list_index = []
for i in range(0, 10):
    list_index.append(i)

data = dataset.iloc[list_index, :]

fig, axs = plt.subplots(1, 1, figsize=(12, 10))
sns.barplot(data=data, x="Department /Ministry", y="Fund allotted(in ₹crores)", ax=axs)
axs.set_xticklabels(axs.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

list_explode = []
for i in range(0, data.shape[0]):
    list_explode.append(0.03)

explode = tuple(list_explode)
fig, axs = plt.subplots(1, 1, figsize=(10, 10))
plt.pie(data[data.columns[1]], labels=data[data.columns[0]], autopct="%.1f%%", pctdistance=0.5,explode=explode)
plt.tight_layout()
plt.show()
```
