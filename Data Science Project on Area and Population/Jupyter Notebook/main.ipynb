```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display
```

```python
cities = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\Kaggle\Data Science Project on Area and Population\Dataset\california_cities.csv")

display(cities.head())
```

```python
print("Shape of cities = %d Baris & %d Kolom"%cities.shape)
print("Informasi dataset : ")
display(cities.info())
```

```python
fig, axs = plt.subplots(1, 1, figsize=(15, 5))
sns.scatterplot(data=cities, x="longd", y="latd", c=np.log10(cities["population_total"]), cmap="viridis", s=cities["area_total_km2"], ax=axs)
axs.set_title("Scatter Plot Californis Citizen", fontsize=20, fontweight="bold")
axs.set_xlabel("Longitude", fontsize=15)
axs.set_ylabel("Latitude", fontsize=15)
axs.grid()
plt.tight_layout()
plt.show()
```
