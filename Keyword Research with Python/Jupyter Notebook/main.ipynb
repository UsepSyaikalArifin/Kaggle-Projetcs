```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from pytrends.request import TrendReq
from IPython.display import display
```

```python
trend = TrendReq()
```

```python
trend.build_payload(kw_list=["Data Science"])
data = trend.interest_by_region().reset_index()
data = data.sort_values(by="Data Science", ascending=False)
display(data)
```

```python
new_data = data[data["Data Science"] >= 10].reset_index(drop=True)
display(new_data.head(5))
```

```python
fig, axs = plt.subplots(1, 1, figsize=(15, 8))
sns.barplot(data=new_data, x="geoName", y="Data Science", palette="viridis", ax=axs)
axs.set_title("Komparasi Ketertarikan Terhadap Data Science\nBerdasarkan Negara", fontsize=20, fontweight='bold')
axs.set_xticklabels(axs.get_xticklabels(), rotation=80)
axs.set_xlabel("Geo Name", fontsize=15)
axs.set_ylabel("Data Science", fontsize=15)
plt.tight_layout()
plt.show()
```
